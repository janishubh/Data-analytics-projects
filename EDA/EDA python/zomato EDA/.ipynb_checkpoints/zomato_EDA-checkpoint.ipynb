```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv',encoding = 'latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
[i for i in df.columns if df[i].isnull().sum()>0]
```

```python
df_country = pd.read_excel('country-code.xlsx')
df_country
```

```python
data = pd.merge(df,df_country,on='Country Code',how='left')
data.head()
```

```python
df.shape
```

```python
data.shape
```

```python
data.info()
```

```python
data.columns
```

```python
country_names = data.Country.value_counts().index
country_names
```

```python
country_val = data.Country.value_counts().values
country_val
```

```python
#pie chart - top 3 countries using zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observati Most of the zomato records in this data set are from India followed by some from USA and UK

```python
ratings = data.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
plt.rcParams['figure.figsize'] = (12,7)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

# Observation:
    1. Not rated count is very high
    2. Most of ratings are between 2.9 to 3.9 
    

```python
#countries that have given 0 rating
data.groupby(['Aggregate rating','Country']).size().reset_index().rename(columns={0:'RatingCount'}).head(5)
```

# Observation:
    Maximum 0 ratings are from Indian customers

```python
#countries and their currency

data[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```
